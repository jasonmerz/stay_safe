```python
import pandas as pd
from config import api_key
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('stay_safe/Crime Map Offense Report_2024-02-20T21_11_06-06_00.xlsx', skiprows = 2)
```

```python
import pandas as pd
import requests
import time

# Assuming df is your DataFrame and it contains an 'Address' column
# Creating placeholder columns for Latitude and Longitude
df['Latitude'] = None
df['Longitude'] = None

def geocode_address(address, api_key):
    base_url = "https://geocode.maps.co/search"
    params = {
        'q': address,
        'api_key': api_key
    }
    response = requests.get(base_url, params=params)
    if response.status_code == 200:
        data = response.json()
        if data:
            return data[0]['lat'], data[0]['lon']
    return None, None

# Initialize a counter
counter = 0

for index, row in df.iterrows():
    try:
        lat, lon = geocode_address(row['Address'], api_key)  # Ensure api_key is defined somewhere in your code
        if lat and lon:
            df.at[index, 'Latitude'] = lat
            df.at[index, 'Longitude'] = lon
            counter += 1  # Increment the counter
        time.sleep(1.2)  # Respect the API rate limit
    except Exception as e:
        print(f"Error geocoding address {row['Address']}: {e}")
    
    # Print the counter every time an address is processed
    print(f"Processed {counter} records")

# Optionally, print the final count after the loop
print(f"Total records processed: {counter}")
```

### Data Preperation and Processing

```python
# Standardizing column names 
df.columns = [str.lower(e).replace(' ','_') for e in df.columns]
```

```python
# Dropping indistinguishable, or non-applicable ZIP Codes
df = df.loc[df['zip_code'].notnull()]
```

```python
# Changing zipcode type to int vs. float
df.zip_code = df.zip_code.astype(int)
```

```python
# Converting String Date format to Actual Date format
df['incident_date'] = pd.to_datetime(df['incident_date'], format='%a %b %d %Y')
```

```python
# Adding year month vector
df['year_month'] = pd.to_datetime((df.incident_date.dt.year *100 + df.incident_date.dt.month).astype(str),  format='%Y%m')
```

```python
df.drop('month_year', axis = 1, inplace = True)
```

```python
# Storing data for future use
df.to_csv('crime_map_data.csv', index = None)
```
